```python
import numpy as np
import pandas as pd

import sys 
#!{sys.executable} -m pip install seaborn --user

import seaborn as sns
import matplotlib.pyplot as plt 

# Input data files are available in the "../input/" directory.
# For example, running this (by clicking run or pressing Shift+Enter) will list the files in the input directory

import os
print(os.listdir("./input"))

# Any results you write to the current directory are saved as output.
```

```python
data = pd.read_csv('./input/Churn_Modelling.csv')
```

## АНАЛИЗ ДАННЫХ

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['RowNumber', 'CustomerId', 'Surname', 'Geography'], axis=1, inplace=True)
```

```python
data.Gender = [1 if each == 'Male' else 0 for each in data.Gender]
```

```python
data.sample(5)
```

## РАЗДЕЛЕНИЕ ПРИЗНАКОВ И ЦЕЛЕЙ

```python
data.Exited.describe()
```

```python
y = data.Exited.values
x_data = data.drop(['Exited'], axis=1)
```

```python
x_data.describe()
```

## НОРМАЛИЗАЦИЯ

```python
x = (x_data - np.min(x_data)) / (np.max(x_data)-np.min(x_data))
x.head()
```

## РАЗБИВКА ДАННЫХ ДЛЯ ОБУЧЕНИЯ И ТЕСТИРОВАНИЯ

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.10, random_state=7)
```

```python
print('x_train shape: ', x_train.shape)
print('y_train shape: ', y_train.shape)
print('x_test shape: ', x_test.shape)
print('y_test shape: ', y_test.shape)
```

## ОБУЧЕНИЕ

```python
from sklearn.utils import class_weight

class_weights = class_weight.compute_class_weight(class_weight = 'balanced',
                                                 classes = np.unique(y_train),
                                                  y = y_train)

class_weights_dict = {}
for i, v in enumerate(class_weights):
    class_weights_dict[i] = v
class_weights
```

- ### KNN

```python
from sklearn.model_selection import train_test_split,GridSearchCV,cross_val_score
cv_result = {}
best_estimators = {}

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report

# name = "knn"
# knn_params= {"n_neighbors": np.linspace(1,19,10, dtype = int).tolist(),
#                  "weights": ["uniform","distance"],
#                  "metric":["euclidean","manhattan"]}

# clf = GridSearchCV(KNeighborsClassifier(), param_grid=knn_params, cv =10, n_jobs = -1,verbose = False)
# clf.fit(x_train,y_train)
# cv_result[name]=clf.best_score_

# best_estimators[name]=clf.best_estimator_
# print(name,'cross validation accuracy : %.3f'%cv_result[name])

# print(clf.best_params_)
# y_pred_knn = clf.predict(x_test)
# # print classification report 
# print(classification_report(y_test, clf_predictions)) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=9, weights='distance') #weights='distance'

knn.fit(x_train, y_train)

y_pred_knn = knn.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

knn_cm = confusion_matrix(y_test, y_pred_knn)

# Visualization:
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(knn_cm, annot=True, linewidth=0.7, linecolor='cyan', fmt='.0f', ax=ax, cmap='BrBG')
plt.title('KNN Classification Confusion Matrix')
plt.xlabel('y_pred')
plt.ylabel('y_test')
plt.show()
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
# score_knn = knn.score(x_test, y_test)
knn_accuracy = accuracy_score(y_test, y_pred_knn)
knn_precision = precision_score(y_test, y_pred_knn, average='macro')
knn_recall = recall_score(y_test, y_pred_knn, average='macro')
knn_f1 = f1_score(y_test, y_pred_knn, average='macro')
knn_roc_auc = roc_auc_score(y_test, knn.predict_proba(x_test)[:, 1])

# print(score_knn)
print("knn_accuracy: ", knn_accuracy)
print("knn_precision: ", knn_precision)
print("knn_recall: ", knn_recall)
print("knn_f1: ", knn_f1)
print("knn_roc_auc: ", knn_roc_auc)
```

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(knn, x_test, y_test)
```

- ### SVM

```python
from sklearn.svm import SVC

svm = SVC(gamma='scale', random_state=1, probability=True, class_weight='balanced') #{0:1, 1:3.1}

svm.fit(x_train, y_train)

y_pred_svm = svm.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

svm_cm = confusion_matrix(y_test, y_pred_svm)

# Visualization:
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(svm_cm, annot=True, linewidth=0.7, linecolor='cyan', fmt='.0f', ax=ax, cmap='BrBG')
plt.title('SVM Classification Confusion Matrix')
plt.xlabel('y_pred')
plt.ylabel('y_test')
plt.show()
```

```python
# score_svm = svm.score(x_test, y_test)
# print(score_svm)

svm_accuracy = accuracy_score(y_test, y_pred_svm)
svm_precision = precision_score(y_test, y_pred_svm, average='macro')
svm_recall = recall_score(y_test, y_pred_svm, average='macro')
svm_f1 = f1_score(y_test, y_pred_svm, average='macro')
svm_roc_auc = roc_auc_score(y_test, svm.predict_proba(x_test)[:, 1])

print("svm_accuracy: ", svm_accuracy)
print("svm_precision: ", svm_precision)
print("svm_recall: ", svm_recall)
print("svm_f1: ", svm_f1)
print("svm_roc_auc: ", svm_roc_auc)
```

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(svm, x_test, y_test)
```

- ### Ансамбль

```python
from sklearn.ensemble import VotingClassifier

ensemble = VotingClassifier(estimators=[('rf', rf), ('svm', svm), ('knn', knn), ('lgb', lgb)], voting='soft')

ensemble.fit(x_train, y_train)

y_pred_ensemble = ensemble.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

ensemble_cm = confusion_matrix(y_test, y_pred_ensemble)

# Visualization:
f, ax = plt.subplots(figsize=(5,5))
sns.heatmap(ensemble_cm, annot=True, linewidth=0.7, linecolor='cyan', fmt='.0f', ax=ax, cmap='BrBG')
plt.title('Ensemble Classification Confusion Matrix')
plt.xlabel('y_pred')
plt.ylabel('y_test')
plt.show()
```

```python
ensemble_accuracy = accuracy_score(y_test, y_pred_ensemble)
ensemble_precision = precision_score(y_test, y_pred_ensemble, average='macro')
ensemble_recall = recall_score(y_test, y_pred_ensemble, average='macro')
ensemble_f1 = f1_score(y_test, y_pred_ensemble, average='macro')
ensemble_roc_auc = roc_auc_score(y_test, ensemble.predict_proba(x_test)[:, 1])

print("ensemble_accuracy: ",ensemble_accuracy)
print("ensemble_precision: ",ensemble_precision)
print("ensemble_recall: ",ensemble_recall)
print("ensemble_f1: ",ensemble_f1)
print("ensemble_roc_auc: ",ensemble_roc_auc)
```

```python
from sklearn.metrics import RocCurveDisplay
RocCurveDisplay.from_estimator(ensemble, x_test, y_test)
```
